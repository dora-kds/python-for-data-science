# VI. Множини (Set).

17. Створити множину без елементів. Після цого за допомогою методу додайте кілька різних елементів до множини. Чи множини є змінним типом даних?

```python
my_set = set()
my_set.add(1)
my_set.add("lil")
my_set.add("hello")
my_set.add(2.34)

print(my_set)
```

Множини є змінним типом даних тому, що ми можемо додавати/видаляти елементи вже після створення сету.

18. Створити множину, маючи список my_list = [1, 1, 2, 67, 67, 8, 9]. Пояснити, чому "зникли" деякі елементи.

```python
my_list = [1, 1, 2, 67, 67, 8, 9]
my_list_set = set(my_list)
print(my_list_set)
```

Після створення сету зникло одне з чисел 67, яке було у листі. Це відбулося через особливість структури множин, в яких дублікати автоматично видаляються.

19. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
set1 = {1, 2, 3, 44, 5}
set2 = {4, 1, 33, 5, 3}

set_union = set1.union(set2)
print(set_union)
set_diff = set1.difference(set2)
print(set_diff)
set_intersection = set1.intersection(set2)
print(set_intersection)
set_simmtr = set1.symmetric_difference(set2)
print(set_simmtr)
```
