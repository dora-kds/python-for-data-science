6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
import math

class Cylinder(Figure_3D):
  def __init__(self, radius, height):
    self.radius = radius
    self.height = height

  def get_volume(self):
    return math.pi * (self.radius**2) * self.height

class Cube(Figure_3D):
  def __init__(self, side):
    self.side = side

  def get_volume(self):
    return self.side**3

cube = Cube(5)
cylinder = Cylinder(3, 7)
total_volume = cube + cylinder
print(total_volume)
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
  def __init__(self, id, balance=0):
    self.__balance = balance
    self.__id = id

  @property
  def balance(self):
    return self.__balance

  @property
  def id(self):
    return self.__id

  def replenishment(self, amount):
    if amount > 0:
      self.__balance += amount
      print(f"Operation is successful! New balance: {self.__balance}")
    else:
      print("Amount has to be more than 0")

  def withdrawal(self, amount):
    if amount <= 0:
      print("Amount has to be more than 0")
    elif amount > self.__balance:
      print("There is no enought money on balance")
    else:
      self.__balance -= amount
      print(f"Operation is successful! New balance: {self.__balance}")

OlegWallet = EWallet(92837, 64)
print(OlegWallet.balance , OlegWallet.id)

OlegWallet.replenishment(50)
OlegWallet.replenishment(0)
OlegWallet.withdrawal(30)
OlegWallet.withdrawal(200)
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
  def __init__(self, name, price):
    self.name = name
    self.price = price

  def __str__(self):
    return f"{self.name} for {self.price} grn"

class ShoppingCart:
  def __init__(self):
    self.items = []

  def add_item(self, item):
    self.items.append(item)
    print(f"{item.name} added to Shopping Cart")

  def remove_item(self, item_name):
    for item in self.items:
      if item.name == item_name:
        self.items.remove(item)
        print(f"{item_name} is successfully removed")
        return
    print(f"{item_name} is not in a Shopping Cart")

  def showing(self):
    if not self.items:
      print("Shopping Cart is empty")
    else:
      for item in self.items:
        print(item)

MilkaCart = ShoppingCart()
item1 = Item("Milk", 60)
item2 = Item("Bread", 40)
item3 = Item("Seafood", 220)

MilkaCart.add_item(item1)
MilkaCart.add_item(item2)
MilkaCart.add_item(item3)

MilkaCart.showing()
MilkaCart.remove_item("Bread")
MilkaCart.showing()
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class Weather:
  def __init__(self):
    self.days = ["sunny", "snowy", "rainy", "frosty", "foggy", "cloudy", "sunny"]
    self.index = 0

  def __iter__(self):
    return self

  def __next__(self):
        if self.index < len(self.days):
          day = self.days[self.index]
          self.index +=1
          return day
        else:
          raise StopIteration

weather_for_week = Weather()
for day in weather_for_week:
  print(day)
```
